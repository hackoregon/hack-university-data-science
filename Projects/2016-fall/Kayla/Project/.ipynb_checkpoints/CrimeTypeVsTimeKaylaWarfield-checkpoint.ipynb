```python
police2012 =pd.read_csv("NIJ2012_MAR01_DEC31.csv", parse_dates=['occ_date'])
police2013 = pd.read_csv("NIJ2013_JAN01_DEC31.csv", parse_dates=['occ_date'])
police2014 = pd.read_csv("NIJ2014_JAN01_DEC31.csv", parse_dates=['occ_date'])
police2015 = pd.read_csv("NIJ2015_JAN01_DEC31.csv", parse_dates=['occ_date'])
jan = pd.read_csv("NIJ2016_JAN01_JUL31.csv", parse_dates=['occ_date'])
aug = pd.read_csv("NIJ2016_AUG01_AUG31_USE.csv", parse_dates=['occ_date'])
sep = pd.read_csv("NIJ2016_SEP01_SEP30.csv", parse_dates=['occ_date'])
octo = pd.read_csv("NIJ2016_OCT01_OCT31.csv", parse_dates=['occ_date'])
```

```python
police2013.tail()
```

```python
refdate = pd.to_datetime("2012-03-01")
def weeknum(date):
    diff=pd.to_datetime(date)-refdate
    return int(diff.days/7)
def monthnum(date):
    diff=pd.to_datetime(date)-refdate
    return int(diff.days/30)
def yearnum(date):
    diff=pd.to_datetime(date)-refdate
    return int(diff.days/132)
```

```python
police2013['week'] = police2013['occ_date'].apply(lambda x: weeknum(x))
police2014['week'] = police2014['occ_date'].apply(lambda x: weeknum(x))
police2015['week'] = police2015['occ_date'].apply(lambda x: weeknum(x))
#police2015['month'] = police2015['occ_date'].apply(lambda x: x.strftime('%m'))

police_all = police2013.append(police2014.append(police2015.append(police2012.append(jan.append(aug.append(sep.append(octo)))))))
police_all['week']= police_all['occ_date'].apply(lambda x: weeknum(x))
police_all['month']= police_all['occ_date'].apply(lambda x: monthnum(x))
police_all['year']= police_all['occ_date'].apply(lambda x: yearnum(x))
```
